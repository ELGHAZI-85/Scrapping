```python
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By

# Démarrez le pilote Web et chargez la page
driver = webdriver.Chrome()
driver.get("https://twitter.com/search?q=cristiano%20ronaldo&src=typed_query")

# Initialiser une liste vide pour stocker les tweets, likes, user_tags, nombre de replys, nombre de retweets et le timestamp


UserTags=[]
Tweets=[]
Replys=[]
reTweets=[]
Likes=[]

# Faites défiler la page et chargez plus de tweets jusqu'à ce que nous ayons 10000 tweets

while len(Tweets) < 5000:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    # Obtenir la source de la page mise à jour
    html = driver.page_source

    # Créer un objet BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Retrouvez tous les éléments du tweet
    tweets = soup.find_all("article",{"data-testid":'tweet'})
    
   
    # Parcourez les tweets et extrayez les données qui nous intéressent

    for tweet in tweets:
        #Extraction du nom d'utilisateur
        try:
            UserTag = tweet.find('span', class_='css-901oao css-16my406 css-1hf3ou5 r-poiln3 r-bcqeeo r-qvutc0').text
            UserTags.append(UserTag)
        except:
            UserTags.append('')
        print(UserTag)
        #Extraction du texte du tweet
        try:
            Tweet = tweet.find('div', class_='css-901oao r-18jsvk2 r-37j5jr r-a023e6 r-16dba41 r-rjixqe r-bcqeeo r-bnwqim r-qvutc0').text
            Tweets.append(Tweet)
        except:
            Tweets.append('')

        #Extraction du timestamp de publication du tweet
        """
        try:
            TimeStamp = tweet.find('time')['datetime']
            TimeStamps.append(TimeStamp)
        except:
            TimeStamps.append('')
        
        """
        #Extraction du nombre de réponse
        try:
            Reply = driver.find_element(By.XPATH,".//div[@data-testid='reply']").text
            Replys.append(Reply)
        except:
            Replys.append('')
       
        #Extraction du nombre de retweet
        try:    
            reTweet = driver.find_element(By.XPATH,".//div[@data-testid='retweet']").text
            reTweets.append(reTweet)
        except:
            reTweets.append('')
        
        #Extraction du nombre de Likes
        try:
            Like = driver.find_element(By.XPATH,".//div[@data-testid='like']").text
            Likes.append(Like)
        except:
            Likes.append('')
       
        
        

# Fermeture webdriver
driver.quit()

#Sauvegarder les données collectées dans une dataframe.
df = pd.DataFrame(zip(UserTags,TimeStamps,Tweets,Replys,reTweets,Likes) ,columns=['UserTags','TimeStamps','Tweets','Replys','reTweets','Likes'])
print(df)
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('wordnet')
nltk.download('sentiwordnet')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
lemmatizer = WordNetLemmatizer()
```

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

def remove_punctuation(text):
    for x in text:
        if x in punctuations:
            text = text.replace(x, "")
    return text

df['Tweets'] = df['Tweets'].apply(remove_punctuation)
```

```python
import re 
#preprocessing
def tweet_preprocess(raw_tweet):
    raw_tweet = re.sub("@\w+","",raw_tweet).strip()
    raw_tweet = re.sub("http\S+","",raw_tweet).strip() #Delete special caracters
    letters_only = re.sub("[^a-zA-Z]", " ",raw_tweet) 
    words = letters_only.lower().split()     
    
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops] 
    lemma = [lemmatizer.lemmatize(word) for word in meaningful_words]
    return lemma#( " ".join( meaningful_words )) 
```

```python
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
```

```python
def get_sentiment(word,tag):
    wn_tag = penn_to_wn(tag) 
    if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):
        return []
    
    synsets = wn.synsets(word, pos=wn_tag)
    if not synsets:
        return []

    # Take the first sense, the most common
    synset = synsets[0]
    swn_synset = swn.senti_synset(synset.name())

    return [synset.name(), swn_synset.pos_score(),swn_synset.neg_score(),swn_synset.obj_score()]
```

```python
t = 0
for tweet in new_data:
    pos=neg=obj=count=0
    pos_val = nltk.pos_tag(tweet)
    #print(pos_val)
    senti_val = [get_sentiment(x,y) for (x,y) in pos_val]
    #print(senti_val)
    for i in range(len(senti_val)):
      try:
        pos = pos + senti_val[i][1]
        neg = neg + senti_val[i][2]
        obj = obj + senti_val[i][3]

      except:
        continue
    t+=1
    print(" Tweet ", t)
    print("Sentiment of the statement is {0} ".format(pos - neg))
```

```python
#Analyse des sentiments

import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification

# Charger le modèle RoBERTa et le tokeniseur
model = RobertaForSequenceClassification.from_pretrained("roberta-base")
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")

# Fonction pour extraire les sentiments du texte en utilisant RoBERTa
def extract_sentiments(text):
    # Tokeniser le texte de l'avis
    input_ids = torch.tensor(tokenizer.encode(text, add_special_tokens=True)).unsqueeze(0)

    # Prédire les sentiments du texte en utilisant RoBERTa
    with torch.no_grad():
        output = model(input_ids)
    sentiment = output[0].argmax().item()
   
    
    # Retourner "negatif" pour un sentiment négatif, "positif" pour un sentiment positif
    if sentiment == 0:
        return "negatif"
    else:
        return "positif"

# Parcourir chaque avis dans le jeu de données
df['sentiments'] = ""
for i in range(0,len(df['Tweets'])):
    df['sentiments'][i] = extract_sentiments(df['Tweets'][i])

    
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
# Prétraiter les données textuelles
def preprocess(text):
    # Supprimer les mots vides et la ponctuation
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english') and token.isalpha()]

    # Lemmatiser les jetons
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    return lemmatized_tokens

# Créer une matrice de termes de document
vectorizer = CountVectorizer(tokenizer=preprocess)
X = vectorizer.fit_transform(df['Tweets'])

# Monter un modèle LDA
lda = LatentDirichletAllocation(n_components=5)
lda.fit(X)

# Transformer la matrice document-terme en une matrice sujet-terme
topic_term_matrix = lda.transform(X)

# Trouver les principaux sujets pour chaque document
top_topics = topic_term_matrix.argmax(axis=1)
print(top_topics)

# Imprimer les mots les plus importants pour chaque sujet
for topic_idx, topic in enumerate(lda.components_):
    print(f'Topic {topic_idx}:')
    print([vectorizer.get_feature_names()[i] for i in topic.argsort()[:-5:-1]])
```
